```python
import joblib
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import os

os.chdir(r"C:\Users\muhammad\Desktop\ir")

# تحميل البيانات باستخدام مسار صحيح
data = joblib.load("data/vectorizers/tfidf_antique_vectors.joblib")
vectors = data['matrix']

inertia = []
k_range = range(1, 15)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(vectors)
    inertia.append(kmeans.inertia_)

plt.plot(k_range, inertia, 'bx-')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('Elbow method (TF-IDF)')
plt.show()
```
